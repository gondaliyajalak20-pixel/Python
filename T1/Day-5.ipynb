```python
import pandas as pd
import numpy as np
data={"department":["IT","IT","HR","finance","HR","IT"],
       "Employee":["A","B","C","D","E","F"],
        "Salary":[50000,60000,65000,45000,70000,70000],
         "Experience":[2,4,3,1,4,5]}
df=pd.DataFrame(data)
df.groupby('department')
print(df)
df.groupby('department').mean(numeric_only=True)
df.groupby(['Salary','Experience']).sum()
```

```python
df.groupby('department',as_index=True).sum()
```

```python
df.groupby('department',sort=True).sum()
```

```python
df.groupby('department')['Salary'].sum()
```

```python
df.groupby('department').agg({'Salary':['mean','max'], 'Experience':['sum','mean']})
```

```python
df.groupby('department')['Salary'].apply(lambda x:x.max()-x.min())
```
