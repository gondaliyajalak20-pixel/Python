```python
import numpy as np
df=pd.DataFrame({'name':['A','B','C','D','E'],'Department':['cs','it','ec','cs','it'],'score':[60,45,78,90,34]})
df['Grade']=np.where(df['score']>60,'A',np.where(df['score']>40,'B','C'))
df
```

```python
from sklearn.model_selection import train_test_split
df=pd.DataFrame({'Hours':[1,2,3,4,5],'score':[10,45,34,67,80]})
print(df)
x=df[['Hours']]
y=df[['score']]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
print(x_train)
print(x_test)
```
