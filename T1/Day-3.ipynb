```python
import pandas as pd
import numpy as np
df=pd.read_csv('titanic.csv')

```

# Data Cleaning

```python
print("Is Null and isna")
print(df.isnull())
print("---------------------------------------------------------------------------------------------------")

print(df.isna())
print("---------------------------------------------------------------------------------------------------")

print(df.isnull().sum())
print("---------------------------------------------------------------------------------------------------")


      
```

```python
#fillna
#using constants
# using mean,median,mode
# using backward bfill and forward ffill
print("---------------------------------------------------------------------------------------------------------------")
print(df.fillna('130'))
print("---------------------------------------------------------------------------------------------------------------")
df=df.fillna(df.mean())
print(df)
          
```

```python
data={
    'Id':[1,2,3,4,4],
    'Name':['Amit','Priya',np.nan,'Rahul',np.nan],
    'Marks':[80,np.nan,75,85,np.nan],
    'City':['Ahmedabad','rajkot',np.nan,'surat','Rajkot']
}
df=pd.DataFrame(data)

print("-----------------------------------------------------------------------------------------------")
print(df)
print("--------------------------------------------")

```

```python
df=df.fillna(df.mode().iloc[0])
print(df)
```

```python
df=df.median(numeric_only=True)
print(df)
```

```python
df=df.fillna(method='ffill')
print(df)
```

```python
df=df.fillna(method='bfill')
print(df)
```

```python
print("---------------------------------------------------------------------------------------------------------------")
print(df.fillna('130'))
print("---------------------------------------------------------------------------------------------------------------")
df=df.fillna(df.mean())
print(df)
print(df.isnull().sum())
          
```

```python
df['age']=df['age'].fillna(df['age'].mean())
print(df.isnull().sum())#to see if the null values got filled or not
```

```python
df['fare']=df['fare'].fillna(df['fare'].median())
print(df.isnull().sum())
```

```python
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
print(df.isnull().sum())

```

```python
df['cabin']=df['cabin'].fillna(method='bfill')
print(df.isnull().sum())
```

```python
#From our dataset
print(df)
print("----------------------------------------------------")
df=df.dropna()
print(df)
print("------------------------------------------------------")
```

```python
#It will drop any row or column if has nan
df=df.dropna(how='any')
print(df)


```

```python
df=df.dropna(how='all')
print(df)
```

```python
df=df.dropna(how='any',axis=1)
print(df)
```

```python
#all will only drop if whole row(axis=0) or columns(axis=1) value is null
df=df.dropna(how='all',axis=1)
print(df)
```
