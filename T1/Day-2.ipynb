```python
import pandas as pd
import numpy as np
data={
    'Name':['Dhoni','Rohit','Virat','Gill'],
    'Run':[12000,11000,3000,4200],
    'Average':[59.5,48.7,51.2,45.3]
}
df=pd.DataFrame(data,index=['A','B','C','D'])
print(df)
print("---------------------------------------")
print(df['Run'])
print("---------------------------------------")
print(df[['Run','Average']])
print("---------------------------------------")
print("Shape",df.shape)
print("---------------------------------------")
print("Dimension",df.ndim)
print("---------------------------------------")
print("Index",df.index)
print("---------------------------------------")
print("Columns",df.columns)
print("---------------------------------------")
print("Printing Only B to D")
print(df.loc['B'::])
print("---------------------------------------")
#Changing values 
df.loc['C','Run']=4000
print(df)
```

```python
print("Set index Without Drop ")
print(df.set_index('Name'))
print("----------------------------------------------------")
print("Set index with Drop(false)")
print(df.set_index('Name',drop=False))
print(df.set_index(['Name','Average'],drop=True))
print("----------------------------------------------------")
print("Reset Index ")
print(df.reset_index())
```

```python
print("Print this gives true false value ")
print(df['Run']>1000)
print("-------------------------------------------------------------------")
print("This will give values greator than 1000 in form of df")
print(df[df['Run']>1000])
print("-------------------------------------------------------------------")
print("To Print Run >1000 and Average >50")
print(df[(df['Run']>1000) & (df['Average']>50)])
print("-------------------------------------------------------------------")
print("to get Run")
print(df['Run'])
print("-------------------------------------------------------------------")
print("To get Run and Average columns")
print(df[['Run','Average']])
print("-------------------------------------------------------------------")
print("Sorting in Ascending by average")
print(df.sort_values(by='Average'))#not necessary to write by
print("-------------------------------------------------------------------")
print("Sorting in Descending by average")
print(df.sort_values('Average',ascending=False))
print("-------------------------------------------------------------------")
print("Sorting in Descending by average axis=0 ")
print(df.sort_values('Average',ascending=False,axis=0))#I0C1
print("-------------------------------------------------------------------")
print("Sorting in Descending by average axis=1 ")
print(df)
# print(df.sort_values('Average',ascending=False,axis=1)) throws key error as axis=1 works for rows and our rows are named as A,B,C,D
# print(df.sort_values('A',ascending=False,axis=1)) throws type error multiple datatype (integer,float)in rows  
#while using axis=1 datatype should be same for all rows element
print("-------------------------------------------------------------------")

print("Sorting index axis=0 in ascending order")
print(df.sort_index(axis=0))
print("-------------------------------------------------------------------")
print("Sorting index axis=1")
print(df.sort_index(axis=1))
print("-------------------------------------------------------------------")
print("Sorting index axis=0 in descending order ")
print(df.sort_index(axis=0,ascending=False))










```

```python
print("-----------------------------------------------------------------------------------------")
print("Sorting Multiple Values in ascending and descending at a time")
print(df.sort_values(['Run','Average'],axis=0,ascending=[True,False]))
print("-----------------------------------------------------------------------------------------")
```

```python
print("-----------------------------------------------------------------------------------------")
print("Sorting Multiple Values in ascending and descending at a time with ignore index")
print(df.sort_values(['Run','Average'],axis=0,ascending=[True,False],ignore_index=False))
print(df.sort_values(['Run','Average'],axis=0,ascending=[True,False],ignore_index=True))
print("-----------------------------------------------------------------------------------------")
print("Using Inplace False")
print(df.sort_values(['Run','Average'],axis=0,ascending=[True,False],ignore_index=False,inplace=False))
print("------------------------------------------------------------------------------------------")
print("Using Inplace True")
df.sort_values(['Run','Average'],axis=0,ascending=[True,False],ignore_index=False,inplace=True)
print(df)
print("-------------------------------------------------------------------------------------------")
```
