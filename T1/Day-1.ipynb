```python
data=[35.56,70,85,77,87]
s=pd.Series(data,index=['A','B','C','D','E'],name='numbers',dtype='float')
print(s)
print(s.values)
print(s.index)
print(s.shape)
print(s.size)
print(s.ndim)
print(s.hasnans)
print(s.name)
print(s.dtype)
print(s.is_unique)
```
